```python
# @title ComfyUI专业管理脚本增强版
# @markdown 支持多通道独立穿透和精细化管理

import sys
import os
import subprocess
import threading
import time
import socket
import urllib.request
import random
import string
from http.server import BaseHTTPRequestHandler, HTTPServer
import json
from typing import Optional

# 配置参数
使用谷歌云盘 = True  # @param {type:"boolean"}
自动更新_ComfyUI = True  # @param {type:"boolean"}
工作目录 = '/content/ComfyUI'

云穿透配置 = {
    8188: {  # ComfyUI服务端口
        "名称": "AI工作流服务",
        "注册字段": "节点基础url"
    },
    12590: {  # 控制接口端口
        "名称": "节点控制接口",
        "注册字段": "节点控制url"
    }
}

节点注册接口 = "https://e-make-home-properties.trycloudflare.com/注册节点"
安装扩展组件 = True #@param {type:"boolean"}

# 全局状态管理
系统状态 = {
    "运行中": True,
    "重启锁": False,
    "实例标识": ''.join(random.choices(string.ascii_letters + string.digits, k=6)),
    "服务进程": {
        "ComfyUI": None,
        "云穿透": {port: {"进程": None, "链接": None} for port in 云穿透配置}
    }
}

# 依赖安装
def 初始化环境():
    print("正在配置运行环境...")
    try:
        subprocess.check_call(["pip", "install", "fastapi", "uvicorn", "httpx", "pydantic"])
        !wget -qO cloudflared.deb https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
        !dpkg -i cloudflared.deb >/dev/null 2>&1
    except Exception as 异常:
        print(f"环境初始化失败: {异常}")
        sys.exit(1)

初始化环境()

# API服务部分
from fastapi import FastAPI, APIRouter
from pydantic import BaseModel
import uvicorn
import httpx
import asyncio

接口应用 = FastAPI()
路由模块 = APIRouter()

class 节点注册模型(BaseModel):
    节点名称: str
    基础访问地址: Optional[str] = None
    控制接口地址: Optional[str] = None

@路由模块.get("/系统操作/重启")
async def 系统重启():
    if 系统状态["重启锁"]:
        return {"状态": "失败", "消息": "系统正在重启中"}

    系统状态["重启锁"] = True
    try:
        print("🔄 正在执行系统级重启...")
        系统状态["运行中"] = False

        # 终止所有子进程
        if 系统状态["服务进程"]["ComfyUI"]:
            系统状态["服务进程"]["ComfyUI"].terminate()
        for port in 云穿透配置:
            if 系统状态["服务进程"]["云穿透"][port]["进程"]:
                系统状态["服务进程"]["云穿透"][port]["进程"].terminate()

        # 重置状态
        系统状态["运行中"] = True
        系统状态["实例标识"] = ''.join(random.choices(string.ascii_letters + string.digits, k=6))

        # 重新启动服务
        启动核心服务()
        return {"状态": "成功", "新实例ID": 系统状态["实例标识"]}
    finally:
        系统状态["重启锁"] = False

接口应用.include_router(路由模块)

# 服务管理模块
def 生成云穿透链接(端口):
    """独立管理每个穿透通道的生命周期"""
    def 通道监控任务():
        # 等待端口就绪
        while 系统状态["运行中"]:
            with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as 探测套接字:
                if 探测套接字.connect_ex(('localhost', 端口)) == 0:
                    break
            time.sleep(0.5)

        # 启动穿透进程
        穿透进程 = subprocess.Popen(
            ["cloudflared", "tunnel", "--url", f"http://localhost:{端口}"],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )
        系统状态["服务进程"]["云穿透"][端口]["进程"] = 穿透进程

        # 捕获穿透链接
        for 输出行 in 穿透进程.stderr:
            if "trycloudflare.com" in 输出行:
                穿透链接 = 输出行[输出行.find("http"):].strip()
                系统状态["服务进程"]["云穿透"][端口]["链接"] = 穿透链接
                print(f"🔗 [{云穿透配置[端口]['名称']}] 穿透成功: {穿透链接}")

                # 自动注册节点
                注册数据 = {
                    "节点名称": 系统状态["实例标识"],
                    云穿透配置[端口]["注册字段"]: 穿透链接
                }
                asyncio.run(节点注册服务(注册数据))
                break

        # 保持进程监控
        while 系统状态["运行中"]:
            time.sleep(1)

    threading.Thread(target=通道监控任务, daemon=True).start()

async def 节点注册服务(注册数据):
    """节点服务注册"""
    try:
        async with httpx.AsyncClient(timeout=10) as 客户端:
            响应 = await 客户端.post(节点注册接口, json=注册数据)
            响应.raise_for_status()
            print(f"✅ 节点注册成功: {响应.json()}")
    except Exception as 异常:
        print(f"⚠️ 节点注册异常: {异常}")

def 启动ComfyUI():
    """核心服务启动"""
    print("🚀 正在启动AI工作流引擎...")
    if 使用谷歌云盘:
        from google.colab import drive
        drive.mount('/content/drive')
        os.chdir("/content/drive/MyDrive")

    if not os.path.exists(工作目录):
        !git clone https://github.com/mrnfqrbl/ComfyUI {工作目录}

    os.chdir(工作目录)
    if 自动更新_ComfyUI:
        !git pull

    !pip install -q xformers!=0.0.18 -r requirements.txt \
        --extra-index-url https://download.pytorch.org/whl/cu121 \
        --extra-index-url https://download.pytorch.org/whl/cu118

    if 安装扩展组件:
        !python custom_nodes/ComfyUI-Manager/cm-cli.py restore-dependencies

    系统状态["服务进程"]["ComfyUI"] = subprocess.Popen(
        ["python", "main.py", "--dont-print-server"],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.STDOUT
    )

def 启动接口服务():
    """启动控制接口"""
    print("🌐 启动节点控制接口...")
    uvicorn.run(接口应用, host="0.0.0.0", port=12590)

def 启动核心服务():
    """服务协调启动"""
    # 启动核心引擎
    threading.Thread(target=启动ComfyUI, daemon=True).start()

    # 启动所有穿透通道
    for 端口 in 云穿透配置:
        生成云穿透链接(端口)

    # 启动API服务
    threading.Thread(target=启动接口服务, daemon=True).start()

# 主程序入口
if __name__ == "__main__":
    print(f"🆔 当前节点标识: {系统状态['实例标识']}")
    启动核心服务()

    try:
        while True:  # 主线程保活
            time.sleep(60)
            print(f"💗 系统心跳检测 | 实例ID: {系统状态['实例标识']}")
    except KeyboardInterrupt:
        print("\n🛑 收到关闭信号，正在清理资源...")
        系统状态["运行中"] = False
        sys.exit(0)
```

```python
 # @title ComfyUi 启动（并行依赖安装+时间统计+优化）
import logging
import os
import socket
import subprocess
import sys
import threading
import time
import traceback
from concurrent.futures import ThreadPoolExecutor, as_completed
from logging import LoggerAdapter

!pip install loguru
from loguru import logger
level="INFO" # @param ["INFO", "DEBUG"]
logger.remove()
logger.add(sys.stdout, level=level)
# ----------------------
# 全局控制变量
# ----------------------
更新_ComfyUI = True # @param {type:"boolean"}
安装自定义节点 = True # @param {type:"boolean"}
工作区 = "/content/drive/MyDrive/ComfyUI"
准备工作线程数 = 6

# ----------------------
# 全局状态控制
# ----------------------
shutdown_event = threading.Event()
comfyui_process = None
cloudflared_process = None
线程锁 = threading.Lock()
统计信息 = {}
脚本启动时间 = time.time()


# ----------------------
# 工具函数
# ----------------------
def 记录耗时(函数名):
    def decorator(func):
        def wrapper(*args, **kwargs):
            if shutdown_event.is_set():
                return
            start = time.time()
            try:
                result = func(*args, **kwargs)
            except Exception as e:
                logger.exception(f"函数 {函数名} 执行出错")  # 记录详细堆栈信息
                shutdown_event.set()
                raise  # 重新抛出异常，以便主流程处理
            finally:
                耗时 = time.time() - start
                with 线程锁:
                    统计信息[函数名] = 统计信息.get(函数名, 0) + 耗时
            return result
        return wrapper
    return decorator


def 执行命令(命令, 工作目录=None, check=True):
    """执行 shell 命令"""
    try:
        logger.info(f"执行命令: {命令}")
        result = subprocess.run(
            命令,
            cwd=工作目录,
            check=check,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True
        )
        logger.debug(f"命令输出:\n{result.stdout}")
        return result.stdout
    except subprocess.CalledProcessError as e:
        logger.error(f"命令执行失败: {e}")
        raise  # 重新抛出异常，以便上层处理
    except FileNotFoundError as e:
        logger.error(f"找不到文件或命令: {e}")
        raise  # 重新抛出异常
    except Exception as e:
        logger.exception(f"执行命令时发生未知错误")
        raise  # 重新抛出异常


# ----------------------
# 核心功能模块
# ----------------------
@记录耗时("挂载 Google Drive")
def 挂载_google_drive():
    if shutdown_event.is_set():
        return
    try:
        from google.colab import drive
        drive.mount("/content/drive")
        logger.success("Google Drive 挂载成功")
    except Exception as e:
        logger.exception(f"Google Drive 挂载失败")  # 记录详细堆栈信息
        shutdown_event.set()
        raise  # 重新抛出异常


@记录耗时("设置工作区")
def 设置_工作区(路径):
    if shutdown_event.is_set():
        return
    try:
        os.makedirs(路径, exist_ok=True)
        os.chdir(路径)
        logger.info(f"工作区设置成功: {os.getcwd()}")
    except OSError as e:
        logger.error(f"创建目录 {路径} 失败: {e}")
        shutdown_event.set()
        raise  # 重新抛出异常
    except Exception as e:
        logger.exception(f"设置工作区失败")  # 记录详细堆栈信息
        shutdown_event.set()
        raise  # 重新抛出异常


@记录耗时("代码管理")
def 更新_或_克隆_comfyui(工作区, 仓库地址="https://github.com/mrnfqrbl/ComfyUI"):
    if shutdown_event.is_set():
        return
    try:
        if not os.path.exists(os.path.join(工作区, ".git")):  # 检查是否是 git 仓库
            logger.info("首次克隆仓库...")
            执行命令(
                ["git", "clone", "--depth=1", "--progress", 仓库地址, 工作区]
            )
        else:
            logger.info("更新现有仓库...")
            执行命令(["git", "pull"], 工作目录=工作区)
    except Exception as e:
        logger.exception("更新或克隆 ComfyUI 仓库失败")
        shutdown_event.set()
        raise  # 重新抛出异常


@记录耗时("依赖安装")
def 安装_主依赖(工作区):
    if shutdown_event.is_set():
        return
    try:
        执行命令(
            ["pip", "install", "-r", "requirements.txt"],
            工作目录=工作区
        )
        logger.success("ComfyUI 依赖安装完成")
    except Exception as e:
        logger.exception("安装 ComfyUI 依赖失败")
        shutdown_event.set()
        raise  # 重新抛出异常


全局_依赖列表 = []

def 安装_单个节点依赖(节点路径):
    节点名 = os.path.basename(节点路径)
    req_file = os.path.join(节点路径, "requirements.txt")
    install_file = os.path.join(节点路径, "install.py")

    try:
        if os.path.isfile(req_file):
            logger.info(f"记录 {节点名} 的依赖 (requirements.txt)...")
            with open(req_file, "r", encoding="utf-8") as f:
                依赖项 = [line.strip() for line in f if line.strip() and not line.startswith("#")]
                全局_依赖列表.extend(依赖项)  # 将依赖项添加到全局列表
            logger.info(f"已记录 {节点名} 的依赖")

        elif os.path.isfile(install_file):
            logger.info(f"发现 {节点名} 的 install.py，将单独安装")
            return "installpy", 节点路径  # 返回 install.py 的信息，稍后处理
        else:
            logger.warning(f"节点 {节点名} 缺少 requirements.txt 和 install.py，跳过")
            return None  # 没有依赖文件，直接返回
    except Exception as e:
        logger.error(f"处理节点 {节点名} 依赖失败: {e}")
        raise  # 重新抛出异常，以便上层处理

def 安装_单个依赖(依赖项):
    """
    使用 pip 安装给定的依赖项列表。
    """
    try:
        logger.info(f"开始安装依赖项: {依赖项}")
        subprocess.check_call(
            ["pip", "install", 依赖项],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE
        )
        logger.info(f"依赖项安装完成: {依赖项}")
    except subprocess.CalledProcessError as e:
        logger.error(f"安装依赖项失败: {依赖项} - {e.stderr.decode()}")
        raise

def 安装_installpy(节点路径):
    """
    运行 install.py 文件。
    """
    节点名 = os.path.basename(节点路径)
    try:
        logger.info(f"安装 {节点名} 的依赖 (install.py)...")
        执行命令(
            ["python", "install.py"],
            工作目录=节点路径
        )
        logger.info(f"{节点名} 依赖安装完成 (install.py)")
    except Exception as e:
        logger.error(f"安装 {节点名} 的 install.py 失败: {e}")
        raise

@记录耗时("节点依赖")
def 安装_自定义节点依赖(工作区):
    global 全局_依赖列表  # 声明使用全局变量
    全局_依赖列表 = []  # 重置全局依赖列表
    installpy_任务列表 = [] # 存储 install.py 的任务

    if shutdown_event.is_set():
        return

    try:
        节点目录 = os.path.join(工作区, "custom_nodes")
        if not os.path.exists(节点目录):
            logger.warning(f"custom_nodes 目录不存在: {节点目录}")
            return  # 没有节点，直接返回

        节点列表 = [os.path.join(节点目录, node) for node in os.listdir(节点目录)
                  if os.path.isdir(os.path.join(节点目录, node))]

        # 1. 收集所有节点的依赖
        with ThreadPoolExecutor(max_workers=10) as executor:
            任务列表 = [executor.submit(安装_单个节点依赖, 节点路径) for 节点路径 in 节点列表]

            for future in as_completed(任务列表):
                try:
                    结果 = future.result()
                    if 结果 and 结果[0] == "installpy":
                        installpy_任务列表.append(结果[1]) # 记录 install.py 任务
                except Exception as e:
                    logger.error(f"节点依赖收集任务失败: {e}")
                    shutdown_event.set()
                    # 取消所有其他任务
                    for task in 任务列表:
                        if not task.done():
                            task.cancel()
                    raise  # 重新抛出异常，以便主流程处理

        # 2. 去重依赖项
        去重_依赖列表 = list(set(全局_依赖列表))
        logger.info(f"去重后的依赖列表: {去重_依赖列表}")

        # 3. 并行安装依赖 (requirements.txt)
        with ThreadPoolExecutor(max_workers=15) as executor:  # 调整并发数
            任务列表 = [executor.submit(安装_单个依赖, 依赖) for 依赖 in 去重_依赖列表]

            for future in as_completed(任务列表):
                try:
                    future.result()
                except Exception as e:
                    logger.error(f"依赖安装任务失败: {e}")
                    shutdown_event.set()
                    # 取消所有其他任务
                    for task in 任务列表:
                        if not task.done():
                            task.cancel()
                    raise

        # 4. 并行运行 install.py
        with ThreadPoolExecutor(max_workers=10) as executor:
            任务列表 = [executor.submit(安装_installpy, 节点路径) for 节点路径 in installpy_任务列表]

            for future in as_completed(任务列表):
                try:
                    future.result()
                except Exception as e:
                    logger.error(f"install.py 安装任务失败: {e}")
                    shutdown_event.set()
                    # 取消所有其他任务
                    for task in 任务列表:
                        if not task.done():
                            task.cancel()
                    raise

        logger.success("自定义节点依赖安装完成")

    except Exception as e:
        logger.exception(f"安装自定义节点依赖失败")
        shutdown_event.set()
        raise  # 重新抛出异常

@记录耗时("网络隧道")
def 安装_cloudflared(工作区):
    if shutdown_event.is_set():
        return
    try:
        执行命令(
            ["wget", "-q",
             "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb",
             "-O", "/tmp/cloudflared.deb"]
        )
        执行命令(["dpkg", "-i", "/tmp/cloudflared.deb"])
        logger.success("Cloudflared 安装完成")
    except Exception as e:
        logger.exception(f"Cloudflared 安装失败")
        shutdown_event.set()
        raise  # 重新抛出异常

# ----------------------
# 服务管理模块
# ----------------------
def 启动cloudflared(端口=8188):

  while True:
          time.sleep(0.5)
          套接字 = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
          结果 = 套接字.connect_ex(('127.0.0.1', 端口))
          if 结果 == 0:
            break
          套接字.close()
  logger.info("启动内穿")
  global cloudflared_url  # 使用全局变量存储 cloudflared URL
  cloudflared_url = None
  进程 = subprocess.Popen(["cloudflared", "tunnel", "--url", "http://127.0.0.1:{}".format(端口)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  for 行 in 进程.stderr:
    文本 = 行.decode()
    if "trycloudflare.com " in 文本:
      cloudflared_url = 文本[文本.find("http"):].strip()  # 保存 URL 并去除首尾空格
      logger.info(f"ComfyUI 的访问链接:{cloudflared_url}")
      break
  return 进程
def 运行_comfyui():
    global comfyui_process
    logger.info("启动 ComfyUI")
    try:
        comfyui_process = subprocess.Popen(
            ["python", "main.py", "--dont-print-server"],
            cwd=工作区,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True
        )

        def output_monitor():
            try:
                while not shutdown_event.is_set():
                    line = comfyui_process.stdout.readline()
                    if not line:
                        break
                    line = line.strip()  # 去除行尾空白
                    logger.info(f"ComfyUI: {line}")  # 使用 logger.info
                    if "ERROR" in line:
                        logger.error(line)
                        shutdown_event.set()
                    elif "WARNING" in line:
                        logger.warning(line)
            except Exception as e:
                logger.exception("ComfyUI 输出监控线程出错")  # 记录详细堆栈信息
                shutdown_event.set()

        threading.Thread(target=output_monitor, daemon=True).start()
    except Exception as e:
        logger.exception(f"ComfyUI 启动失败")  # 记录详细堆栈信息
        shutdown_event.set()


# ----------------------
# 主控制流程
# ----------------------
def 清理环境():
    if comfyui_process:
        comfyui_process.terminate()
    if cloudflared_process:
        cloudflared_process.terminate()
    logger.warning("已清理运行环境")


def main():
    try:
        # 第一阶段：关键前置任务
        挂载_google_drive()
        设置_工作区(工作区)

        # 第二阶段：并行初始化
        with ThreadPoolExecutor(max_workers=准备工作线程数) as executor:
            任务列表 = {
                "更新_或_克隆_comfyui": executor.submit(更新_或_克隆_comfyui, 工作区),
                "安装_依赖": executor.submit(安装_主依赖, 工作区),
                "安装_自定义节点依赖": executor.submit(安装_自定义节点依赖, 工作区),
                "安装_cloudflared": executor.submit(安装_cloudflared, 工作区)
            }

            for 任务名, future in 任务列表.items():
                try:
                    future.result()  # 获取结果，如果任务抛出异常，这里会重新抛出
                    logger.info(f"{任务名} 完成")
                except Exception as e:
                    logger.error(f"{任务名} 失败: {e}")
                    shutdown_event.set()
                    # 取消所有其他任务
                    for name, task in 任务列表.items():
                        if task != future and not task.done():
                            task.cancel()
                    break  # 停止后续任务
                finally:
                    # 确保在任务完成后清理资源
                    executor._threads.discard(threading.current_thread())

        if shutdown_event.is_set():
            logger.error("初始化阶段出错，终止程序")
            return

        # 第三阶段：服务启动
        logger.info("启动核心服务...")
        comfyui_thread = threading.Thread(target=运行_comfyui, daemon=True)
        cloudflare_thread = threading.Thread(target=启动cloudflared, daemon=True)

        comfyui_thread.start()
        cloudflare_thread.start()

        # 第四阶段：状态监控
        while not shutdown_event.is_set():
            time.sleep(60)
            with 线程锁:
                状态信息 = "\n".join([f"{k}: {v:.2f}s" for k, v in 统计信息.items()])
                logger.info(f"运行状态监控:\n{状态信息}\n运行时长: {time.time() - 脚本启动时间:.2f}s")

    except Exception as e:
        logger.exception(f"主流程异常")  # 记录详细堆栈信息
        shutdown_event.set()
    finally:
        清理环境()
        sys.exit(1 if shutdown_event.is_set() else 0)


if __name__ == "__main__":
    # 全局异常捕获
    try:
        main()
    except Exception as e:
        logger.exception("全局未捕获异常")  # 记录详细堆栈信息
        清理环境()
        sys.exit(1)
```

```python
# @title 模型下载
!pip install huggingface_hub ipywidgets

from huggingface_hub import snapshot_download
import os
from IPython.display import display
import ipywidgets as widgets

# 🚀 配置镜像（国内用户需要）
# os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"

# 🔧 仓库及文件名定义
MODEL_CONFIG = {
    # 主模型 (GGUF格式)
    "main_model": {
        "repo_id": "city96/FLUX.1-dev-gguf",
        "files": {
            "Q8": "flux-1-dev-Q8_0.gguf",  # 精确文件名
            "Q5": "flux1-dev-Q5_0.gguf"
        },
        "save_path": "/content/models/main"  # 保存目录
    },
    # FLUX专用文本编码器
    "text_encoder": {
        "repo_id": "comfyanonymous/flux_text_encoders",
        "files": {
            "clip_l": "clip_l.safetensors",       # 模型主体
            "t5xxl": "t5xxl_fp8_e4m3fn.safetensors"            # 配置文件
        },
        "save_path": "/content/models/text_encoder"
    },
    # VAE组件
    "vae": {
        "repo_id": "black-forest-labs/FLUX.1-dev",
        "files": {
            "model": "ae.safetensors",
        },
        "save_path": "/content/models/vae"
    }
}

# 🎛️ 创建交互组件
main_q8_check = widgets.Checkbox(description='下载 Q8 主模型 (flux-1-dev-Q8_0.gguf)')
main_q5_check = widgets.Checkbox(description='下载 Q5 主模型 (flux1-dev-Q5_0.gguf)')
textenc_check = widgets.Checkbox(description='下载 FLUX 文本编码器 (clip_l.safetensors, t5xxl_fp8_e4m3fn.safetensors)')
vae_check = widgets.Checkbox(description='下载 VAE ')

status_output = widgets.Output()
download_btn = widgets.Button(description="开始下载", button_style='success')

# 📥 下载函数
def start_download(_):
    with status_output:
        print("🚀 开始精确文件下载...")

        # 下载主模型
        if main_q8_check.value or main_q5_check.value:
            os.makedirs(MODEL_CONFIG["main_model"]["save_path"], exist_ok=True)
            files_to_download = []
            if main_q8_check.value:
                files_to_download.append(MODEL_CONFIG["main_model"]["files"]["Q8"])
            if main_q5_check.value:
                files_to_download.append(MODEL_CONFIG["main_model"]["files"]["Q5"])

            for filename in files_to_download:
                try:
                    print(f"⏬ 正在下载主模型: {filename}")
                    snapshot_download(
                        repo_id=MODEL_CONFIG["main_model"]["repo_id"],
                        allow_patterns=[filename],  # 精确匹配
                        local_dir=MODEL_CONFIG["main_model"]["save_path"],
                        local_dir_use_symlinks=False,  # 强制文件复制
                        resume_download=True
                    )
                except Exception as e:
                    print(f"❌ 主模型 {filename} 下载失败: {str(e)}")

        # 下载文本编码器
        if textenc_check.value:
            os.makedirs(MODEL_CONFIG["text_encoder"]["save_path"], exist_ok=True)
            for filetype, filename in MODEL_CONFIG["text_encoder"]["files"].items():
                try:
                    print(f"⏬ 正在下载文本编码器: {filename}")
                    snapshot_download(
                        repo_id=MODEL_CONFIG["text_encoder"]["repo_id"],
                        allow_patterns=[filename],
                        local_dir=MODEL_CONFIG["text_encoder"]["save_path"],
                        local_dir_use_symlinks=False,
                        resume_download=True
                    )
                except Exception as e:
                    print(f"❌ 文本编码器 {filename} 下载失败: {str(e)}")

        # 下载VAE
        if vae_check.value:
            os.makedirs(MODEL_CONFIG["vae"]["save_path"], exist_ok=True)
            for filetype, filename in MODEL_CONFIG["vae"]["files"].items():
                try:
                    print(f"⏬ 正在下载VAE: {filename}")
                    snapshot_download(
                        repo_id=MODEL_CONFIG["vae"]["repo_id"],
                        allow_patterns=[filename],
                        local_dir=MODEL_CONFIG["vae"]["save_path"],
                        local_dir_use_symlinks=False,
                        resume_download=True
                    )
                except Exception as e:
                    print(f"❌ VAE {filename} 下载失败: {str(e)}")

        print("🎉 所有文件下载完成，目录结构：")
        !tree -h {MODEL_CONFIG["main_model"]["save_path"]}
        !tree -h {MODEL_CONFIG["text_encoder"]["save_path"]}
        !tree -h {MODEL_CONFIG["vae"]["save_path"]}

download_btn.on_click(start_download)

# 🖥️ 显示界面
display(widgets.VBox([
    widgets.HTML("<h3>FLUX 精确文件下载器</h3>"),
    widgets.HTML("<b>主模型选项：</b>"),
    main_q8_check,
    main_q5_check,
    widgets.HTML("<hr>"),
    widgets.HTML("<b>附加组件：</b>"),
    textenc_check,
    vae_check,
    widgets.HTML("<br>"),
    download_btn,
    status_output
]))
```

```python
# @title 从 Civitai API 下载 模型

from google.colab import drive
drive.mount('/content/drive')

模型保存路径 = "/content/drive/MyDrive/ckpt"  # @param {type:"string"}
模型ID = "982559"  # @param {type:"string"}
API密钥 = "301668c17b1b7ffe7c5c0516f323b9db"  # @param {type:"string"}

import os
import requests
import json
from datetime import datetime

# 创建模型保存路径
os.makedirs(模型保存路径, exist_ok=True)

# Civitai API URL (使用模型版本 API)
API_URL = f"https://civitai.com/api/v1/models/{模型ID}"

# 发送 API 请求
headers = {
    "Authorization": f"Bearer {API密钥}"
}
response = requests.get(API_URL, headers=headers)
response.raise_for_status()

# 获取模型信息
model_data = response.json()

# 获取所有模型版本
model_versions = model_data["modelVersions"]

# 按创建日期排序 (从最新到最旧)
model_versions.sort(key=lambda x: datetime.fromisoformat(x["createdAt"].replace("Z", "+00:00")), reverse=True)

# 查找最新版本中的 bf16 文件
download_url = None
模型文件名 = None
for version in model_versions:
    for file in version["files"]:
        if file["type"] == "Model" and file["metadata"] and file["metadata"].get("fp") == "bf16":
            download_url = file["downloadUrl"]
            模型文件名 = file["name"]
            break
    if download_url:
        break
# download_url="https://civitai.com/api/download/models/1499827"
# 模型文件名 ="lightricksLTXV_2b095.safetensors"
if download_url:
    # 构建完整的模型文件路径
    模型文件完整路径 = os.path.join(模型保存路径, 模型文件名)

    print(f"原始下载链接: {download_url}")
    print(f"文件名: {模型文件名}")

    # 设置 User-Agent 和 Referer 头部
    headers = {
        "Authorization": f"Bearer {API密钥}",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",  # 模拟 Chrome 浏览器
        "Referer": "https://civitai.com/"  # 模拟来自 Civitai 域名的请求
    }

    # 使用 HEAD 请求获取重定向后的 URL
    try:
        redirected_response = requests.head(download_url, allow_redirects=True, headers=headers, timeout=10)  # 添加超时时间
        redirected_response.raise_for_status()  # 仅在状态码不是 200-399 时才抛出异常
        final_download_url = redirected_response.url
        print(f"重定向后的下载链接: {final_download_url}")
    except requests.exceptions.RequestException as e:
        print(f"获取重定向 URL 失败: {e}")
        final_download_url = redirected_response.url  # 即使请求失败，也尝试从 response.url 获取 URL
        print(f"尝试从 response.url 获取重定向后的下载链接: {final_download_url}")

    if final_download_url:
        # 使用 aria2 下载模型
        !apt-get install -y aria2
        !aria2c --console-log-level=error -c -x 16 -s 16 -k 1M "{final_download_url}" -d "{模型保存路径}" -o "{模型文件名}"

        print(f"模型已下载到: {模型文件完整路径}")
    else:
        print("无法获取重定向后的下载链接。")
else:
    print("无法找到任何可下载的模型文件。")
    print("请检查模型 ID 是否正确，以及模型是否存在可下载的版本。")
    print("请确保您的 API 密钥具有足够的权限。")
```
